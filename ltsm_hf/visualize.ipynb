```python
import json
import matplotlib.pyplot as plt

root_path = './output/'
file_path = 'ltsm_train_lr0.0005_lora512_down270_freeze0_e500'

with open(root_path+file_path+'/trainer_state.json', 'r') as file:
    data = json.load(file)


log_history = data.get('log_history', [])

filtered_log_history = [entry for entry in log_history if 'loss' in entry]

filtered_epochs = [entry['epoch'] for entry in filtered_log_history]
filtered_losses = [entry['loss'] for entry in filtered_log_history]

plt.figure(figsize=(10, 6))
plt.plot(filtered_epochs, filtered_losses, color='b')
plt.title(file_path)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.savefig(root_path+file_path+'/learning_curve.png')
```
